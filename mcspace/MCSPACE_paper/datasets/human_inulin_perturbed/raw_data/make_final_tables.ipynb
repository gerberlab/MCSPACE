# Clean up data and output processed files for MCSPACE

```python
tempdata = pd.read_csv(datapath / "full_inulin_dataset_for_BWH.csv")
```

```python
tempdata.head()
```

```python
colskeep = ["OTU", "Count", "Particle", "Sample", "Day","Period"]
```

### make tax and perts csv files

```python
taxcols = ['OTU', 'Phylum', 'Order', 'Class', 'Family', 'Genus', 'Species']
```

```python
taxonomy = tempdata[taxcols]
```

```python
taxonomy
```

```python
taxonomy["Domain"] = "Bacteria"
```

```python
taxcolord = ['OTU','Domain', 'Phylum', 'Order', 'Class', 'Family', 'Genus', 'Species']
taxonomy = taxonomy[taxcolord]
```

```python
taxonomy["OTU"] = taxonomy["OTU"].str.replace(r'^OTU', 'Otu', regex=True)
taxonomy.rename(columns={"OTU":"Otu"}, inplace=True)
```

```python
taxonomy=taxonomy.drop_duplicates(subset='Otu')
```

```python
taxonomy = taxonomy.fillna('na')
```

```python
taxonomy
```

```python
# change any 'unclassified' to na as well
taxdf = taxonomy.applymap(lambda x: 'na' if isinstance(x, str) and x.startswith('unclass') else x)
```

```python
taxdf
```

```python
taxdf.to_csv(outpath / "taxonomy.csv")
```

```python
otu_seqs = tempdata[['OTU', '16S_Sequence']]
```

```python
otu_seqs["OTU"] = otu_seqs["OTU"].str.replace(r'^OTU', 'Otu', regex=True)
```

```python
otu_seqs
```

```python
with open(outpath / 'zotus.fa', 'w') as f:
    for index, row in otu_seqs.iterrows():
        header = f">{row['OTU']}"
        sequence = row['16S_Sequence']
        f.write(f"{header}\n{sequence}\n")
```
