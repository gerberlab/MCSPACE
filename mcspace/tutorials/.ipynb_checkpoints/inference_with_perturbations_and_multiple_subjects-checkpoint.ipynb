# Load the dataset

...see also tutorial (***link) for overview of working with datasets

```python
dataset = DataSet(mousepath / "counts.csv",
                 mousepath / "taxonomy.csv")
```

```python
def load_mouse_data(countfile, taxfile):
    dataset = DataSet(countfile, taxfile)
    dataset.remove_subjects(['JX09'])
    dataset.remove_times([43])

    num_consistent_subjects=2
    min_abundance=0.01
    min_reads=1000
    max_reads=10000

    dataset.consistency_filtering(num_consistent_subjects=num_consistent_subjects, 
                                  min_abundance=min_abundance, 
                                  min_reads=min_reads, 
                                  max_reads=max_reads)

    taxa = dataset.get_taxonomy()
    Lactococcus_oidxs = list(taxa.loc[taxa['Genus'] == 'Lactococcus',:].index)
    dataset.filter_particle_data(min_reads=min_reads, max_reads=max_reads)

    reads = dataset.get_reads()
    times = dataset.times
    subjects = dataset.subjects
    taxa = dataset.get_taxonomy()
    return reads, times, subjects, taxa
```

```python
reads, times, subjects, taxonomy = load_mouse_data(mousepath / "counts.csv", mousepath / "taxonomy.csv")
```

```python
num_otus = taxonomy.shape[0]
print(num_otus)
```

# Configure model

```python
# get your local device (option to utilize gpu)
device = get_device()
```

```python
# get data in format for model
data = get_data(reads, device)
```

```python
# model parameters

# TODO: ??add wrappers here to simplify things...
num_assemblages = 100
perturbed_times = [0, 1, -1]  # this is confusing, better to update...
perturbation_prior = 0.5/num_assemblages

sparsity_prior = 0.5/num_assemblages

# better to automate this as well...
num_reads = 0
for t in times:
    for s in subjects:
        num_reads += reads[t][s].sum()
sparsity_prior_power = 0.001*num_reads 

process_var_prior = 0.01
add_process_var=True

```

### add/check learning rate

```python
model = MCSPACE(num_assemblages,
            num_otus,
            times,
            subjects,
            perturbed_times,
            perturbation_prior,
            sparsity_prior,
            sparsity_prior_power,
            process_var_prior,
            device,
            add_process_var,
            use_contamination=True,
            contamination_clusters=data['group_garbage_clusters']
            )
model.to(device)
```

# Run inference

```python
num_epochs = 5000 #5000
elbos = train_model(model, data, num_epochs)
```

```python
fig, ax = plt.subplots()
ax.plot(elbos)
```

## learned assemblages (on tree)

### tools to plot on tree?

```python
import matplotlib.colors as mcolors
```

```python
def add_border(ax, color='black'):
    for _, spine in ax.spines.items():
        spine.set_visible(True)
        spine.set_linewidth(1.0)
        spine.set_color(color)
    return ax
```

```python
def plot_assemblages(ax, theta, logscale=True, vmin=-2, vmax=0, cmap=None, linewidth=0.5, linecolor='#e6e6e6',
                    square=False, cbar=False, xticklabels=True, yticklabels=True):
    if cmap is None:
        cmap = mcolors.LinearSegmentedColormap.from_list('theta cmap', 
                                             [(0,    'white'),
                                              (0.25, 'lightyellow'),
                                              (0.5, 'yellowgreen'),
                                              (1,    'green')], N=256)
    if logscale is True:
        theta_plot = np.log10(theta)
    else:
        theta_plot = theta
        
    ax=sns.heatmap(theta_plot, ax=ax, square=square, cbar=cbar, vmin=vmin, vmax=vmax, cmap=cmap,
                         linewidth=linewidth, linecolor=linecolor, xticklabels=xticklabels, yticklabels=yticklabels)
    return ax
```

```python
fig, ax = plt.subplots()
# ax=vis.plot_assemblages(assemblages) # with default options...
ax=plot_assemblages(ax,assemblages, cbar=True) # with default options...
ax=add_border(ax)
```

```python
def plot_assemblage_proportions(ax, beta, time=None, subject=None, cmap="Blues", logscale=True,
                               vmin=-3, vmax=0, cbar=False, square=False, order=None):
    if time is not None:
        beta = beta.loc[(beta['Time'] == time),:]
    if subject is not None:
        beta = beta.loc[(beta['Subject'] == subject),:]
    betawide = beta.pivot(index='Subject', columns='Assemblage', values="Value") 
    
    if (time is None) and (subject is None):
        raise ValueError("not implemented - please specify at least a time or subject to plot....")
        # to do, if both none, how to order?; what if both not none?, should be ok too...
    
    if order is None:
        ncomm = betawide.shape[1]
        order = [f"A{i+1}" for i in range(ncomm)]
        
    ax=sns.heatmap(betawide.loc[:,order], ax=ax, cmap=cmap, square=square, vmin=vmin, vmax=vmax, cbar=cbar,
           linewidth=linewidth, linecolor=linecolor, xticklabels=xticklabels, yticklabels=yticklabels)
    return ax
```

```python
fig, ax = plt.subplots(nrows=len(times))
for i,t in enumerate(times):
    ax[i]=plot_assemblage_proportions(ax[i], proportions, time=times[i])
    ax[i]=add_border(ax[i])
    ax[i].set_ylabel(f"Day {times[i]}")
```
